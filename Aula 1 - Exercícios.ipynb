```python
df2_filter = df2[df2['nome_drs'].isin(['Grande São Paulo', 'Marília', 'São José do Rio Preto'])].groupby(['nome_drs', 'ano_mes'], as_index=False).sum()[['casos_novos', 'nome_drs', 'ano_mes']]
```

```python
df2_pivot = df2_filter.pivot(index='ano_mes', columns='nome_drs', values='casos_novos')
```

```python
df2_pivot.plot(kind='line', title='Casos de COVID a Cada Mês por DRS', xlabel='Meses', ylabel='Número de Casos', figsize=(15,10))
```
